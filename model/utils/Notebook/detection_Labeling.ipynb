```python
import json
import os
import pandas as pd
import numpy as np
import copy

'''
class_to_label = {
        0 : "Normal",
        10 : "xylostella",
        11 : "white butterfly",
        12 : "striolata",
        15 : "Pentatomidae"
}
'''

class_to_label = {
        0 : 0,
        10 : 1,
        11 : 2,
        12 : 3,
        15 : 4
}



'''
fname = set()
with open('/opt/ml/input/Project/bug_data/Training/train_ann.json') as f:
    json_object = json.load(f)
    
for file in json_object:
    fname.add(file['filename'])

image_file_pathes = '/opt/ml/input/Project/bug_data/Training/image'
image_files = os.listdir(image_file_pathes)

print(len(fname))
print(len(image_files))
count = 0
for file in image_files:
    if file not in fname:
        os.remove(os.path.join(image_file_pathes,file))

image_files = os.listdir(image_file_pathes)

print(len(fname))




'''

json_file_pathes = '/opt/ml/input/Project/bug_data/Training/label' # label json 폴더 경로
json_file_names = os.listdir(json_file_pathes)

labels =[]

for file in json_file_names:
    with open(os.path.join(json_file_pathes,file)) as f:
        json_object = json.load(f)
    
    
    fname = json_object['description']['image']  
    tmp = dict()
    
    tmp['filename'] = fname
    tmp['width'] = 512
    tmp['height'] = 512
    tmp['ann'] = dict()
    
    object_list = json_object['annotations']['object']
    for object_one in object_list:
        tmp['ann']['labels'] =class_to_label[object_one['class']]
        x1 = object_one['points'][0]['xtl']
        y1 = object_one['points'][0]['ytl']
        x2 = object_one['points'][0]['xbr']
        y2 = object_one['points'][0]['ybr']
        tmp['ann']['bboxes'] = [x1,y1,x2,y2]
        
        labels.append(copy.deepcopy(tmp))

        
with open('./train_ann.json', 'w') as outfile:
    json.dump(labels,outfile)
print(labels)

    
    
    
    
```
